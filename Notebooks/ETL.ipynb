```python
def extract_company_names(x):
    try:
        companies = ast.literal_eval(x) if isinstance(x, str) else x
        return [i['name'] for i in companies if isinstance(i, dict) and 'name' in i]
    except:
        return []

# Aplicar la función a la columna
df_movies['production_companies'] = df_movies['production_companies'].apply(extract_company_names)

# Verificar los resultados
print(df_movies['production_companies'].head(10))
```

```python
# Me aseguro de rellenar valores nulos en 'revenue' y 'budget'
df_movies['revenue'] = df_movies['revenue'].fillna(0)
df_movies['budget'] = df_movies['budget'].fillna(0)
```

```python
#Elimino nulos y formateo fechas en 'release_date', ademas creo la columna 'release_year' con el año de lanzamiento
df_movies = df_movies.dropna(subset=['release_date'])
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])
df_movies['release_year'] = df_movies['release_date'].dt.year
```

```python
# Me aseguro que 'revenue' y 'budget' sean de tipo numerico
df_movies['revenue'] = pd.to_numeric(df_movies['revenue'], errors='coerce')
df_movies['budget'] = pd.to_numeric(df_movies['budget'], errors='coerce')
```

```python
#Convierto 'cast' y 'crew' a list
df_credits['cast'] = df_credits['cast'].apply(ast.literal_eval)
df_credits['crew'] = df_credits['crew'].apply(ast.literal_eval)
```

Extraigo los nombres de los actores en una lista

```python
def get_all_actors(cast_list):
    return [actor['name'] for actor in cast_list]

#Aplico la funcion y creo una columna nueva 'actors'
df_credits['actors'] = df_credits['cast'].apply(get_all_actors)
```

```python
print(df_credits['actors'].iloc[0])
```

Extraigo los nombres de los directores

```python
def get_director(crew_list):
    for member in crew_list:
        if member['job'] == 'Director':
            return member['name']
    return None

#Aplico la funcion y creo una nueva columna 'director'
df_credits['director'] = df_credits['crew'].apply(get_director)
```

```python
#Tomo una muestra seleccionando 10000 peliculas aleatorias de df_movies_final
#Elijo el tamaño de la muestra
sample_size = 10000

df_movies_sample = df_movies_final.sample(n=sample_size, random_state=42)
#En el caso de credits, selecciono las filas que corresponden con las peliculas en el sample de df_movies
df_credits_sample = df_credits[df_credits['id'].isin(df_movies_sample['id'])]
```

```python
# Hago el merge de los dataframes 
df_combined = pd.merge(df_movies_sample, df_credits_sample, on='id', how='inner')
```
